```python
def plot(x):
    """
    :return:
    """
    data = x
    fig = plt.figure()
    ax = fig.add_subplot(121, projection='3d')

    _x = np.arange(len(data[0]))
    _y = np.arange(len(data))
    _xx, _yy = np.meshgrid(_x, _y)
    x, y = _xx.ravel(), _yy.ravel()

    topp = data.ravel()
    bottom = np.zeros_like(topp)
    width = depth = 0.50
    
    
    ax.bar3d(*data, bottom, width, depth, topp, shade=True)
    ax.set_title("Plot of Delta")

    fig.set_size_inches(onecolumn_width, onecolumn_height)
    plt.show()
```

```python
#Using this one

def comparison_plot(batched_t, original, name):
    
    n_col = 4
    n_rows = 2
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_col, sharey=True, sharex=True)

    original = np.array(original)
    
    for i in range(n_rows):
        k = 0
        for j in range(n_col):
            
            #Plot original first
            if i is 0:
                showing = original[k]

                if batched_t.shape[-1] == 3:
                    showing = showing
                else:
                    showing = showing.permute(1, 2, 0)
                    
            #Plot adversarial below it        
            else:
                showing = batched_t[k]

                if batched_t.shape[-1] == 3:
                    showing = showing
                else:
                    showing = showing.permute(1, 2, 0)
                    
                    
            axes[i, j].imshow(showing)
            k += 1

    fig.set_size_inches(twocolumn_width, twocolumn_height)
    plt.tight_layout()
    fig.savefig(os.path.join(out_dir, '{}.pdf'.format(name)))
    
    
```

```python
#Script to get L2 norm of deltas, average iterations, average cosine similiarity for a c value

import pickle as pkl
import re
import skvideo.io
from scipy import spatial
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer



c_vals = [0.01, 0.1, 1, 10, 100] 
for c in c_vals:
    pkl_path = 'D:\\College\\Research\\December 2018 Video Captioning Attack\\3212019 results\All c results\\{}\\'.format(c)
    video_path = 'D:\\College\\Research\\December 2018 Video Captioning Attack\\video captioner\\YouTubeClips\\'

    print(pkl_path)
    path_list =  [os.path.join(root, name)
                    for root, dirs, files in os.walk(pkl_path)
                    for name in files
                    if name.endswith((".pkl"))]
    
    #pass_in, iterations, delta, final_caption, original_caption, target_caption
    for path in path_list:
        avg_delta = []
        total_iterations = []
        cos_similarity = []
        

        with open(path, 'rb') as f:
            print("Loaded %s" % path)
            x = pkl.load(f)

        pass_in = x['pass_in']
        delta = x['delta']
        iterations = x['iterations']
        target_caption = x['target_caption'].replace('<sos>', '').replace('<eos>', '')
        final_caption = x['final_caption'].replace('<sos>', '').replace('<eos>', '')
        
        documents = (target_caption, final_caption)
        tfidf_vectorizer=TfidfVectorizer(analyzer="char")
        tfidf_matrix=tfidf_vectorizer.fit_transform(documents).toarray()

        cosine_similarity = 1 - spatial.distance.cosine(tfidf_matrix[0], tfidf_matrix[1])
        cos_similarity.append(cosine_similarity)
        for f_delta in delta:
            avg_delta.append(f_delta[f_delta != 0])
        total_iterations.append(iterations)

    print("C value: {}\nL2 Norm of Delta:{}\n".format(c, np.linalg.norm(avg_delta, 2)))
    print("Average Cosine Similarity: {}\nAverage number of iterations: {}\n".format(np.mean(cos_similarity), np.mean(total_iterations)))
# print(np.mean(avg_delta))
```
