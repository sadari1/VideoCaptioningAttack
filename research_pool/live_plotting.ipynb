```python
def plot(x):
    """
    :return:
    """
    data = x
    fig = plt.figure()
    ax = fig.add_subplot(121, projection='3d')

    _x = np.arange(len(data[0]))
    _y = np.arange(len(data))
    _xx, _yy = np.meshgrid(_x, _y)
    x, y = _xx.ravel(), _yy.ravel()

    topp = data.ravel()
    bottom = np.zeros_like(topp)
    width = depth = 0.50
    
    
    ax.bar3d(*data, bottom, width, depth, topp, shade=True)
    ax.set_title("Plot of Delta")

    fig.set_size_inches(onecolumn_width, onecolumn_height)
    plt.show()
```

```python
#Using this one

def comparison_plot(batched_t, original, name):
    
    n_col = 4
    n_rows = 2
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_col, sharey=True, sharex=True)

    original = np.array(original)
    
    for i in range(n_rows):
        k = 0
        for j in range(n_col):
            
            #Plot original first
            if i is 0:
                showing = original[k]

                if batched_t.shape[-1] == 3:
                    showing = showing
                else:
                    showing = showing.permute(1, 2, 0)
                    
            #Plot adversarial below it        
            else:
                showing = batched_t[k]

                if batched_t.shape[-1] == 3:
                    showing = showing
                else:
                    showing = showing.permute(1, 2, 0)
                    
                    
            axes[i, j].imshow(showing)
            k += 1

    fig.set_size_inches(twocolumn_width, twocolumn_height)
    plt.tight_layout()
    fig.savefig(os.path.join(out_dir, '{}.pdf'.format(name)))
    
    
```
